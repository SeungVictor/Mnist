### Dataload

```python
# 데이터세트 디렉터리 지정
train_dataset_dir = "/Users/spark/OneDrive/ps/Dacon/Mnist/dataset/train/"
test_dataset_dir = "/Users/spark/OneDrive/ps/Dacon/Mnist/dataset/test/"

# csv 파일 로드
train_dataset_csv = pd.read_csv(train_dataset_dir + "train_data.csv")
test_dataset_csv = pd.read_csv(test_dataset_dir + "test_data.csv")
```

```python
# 학습 데이터세트와 데이터 라벨 저장
dataset = []
dataset_label = []

# 학습 데이터세트에서 이미지를 불러와 Array 형태로 저장합니다.
for file in train_dataset_csv['file_name']:
    image_array = np.array(Image.open(train_dataset_dir + file))
    dataset.append(image_array)
    
# 학습 데이터 라벨을 저장합니다.
for label in train_dataset_csv['label']:
    dataset_label.append(label)
```

```python
image2number = []
for file in train_dataset_csv['file_name']:
    image_array = np.array(Image.open(train_dataset_dir + file)).flatten()
    image2number.append(image_array)
```

```python
# sklearn 라이브러리를 사용합니다.
from sklearn.manifold import TSNE

# 3~5분 정도 걸려요!
image_tsne = TSNE(n_components=2).fit_transform(image2number)
```

```python
# 예쁜 그림을 그리기 위해 seaborn 라이브러리를 사용합니다.
import seaborn as sns

plt.figure(figsize=(16,10))
plt.title("Digits T-SNE visualization", fontsize=25)
sns.scatterplot(
    x=image_tsne[:,0], y=image_tsne[:,1],
    hue=dataset_label,
    palette=sns.color_palette("hls", 10),
    legend="full"
)
plt.show()
```

```python
# 학습 데이터세트를 List에서 Array 형식으로 변환합니다.
dataset = np.array(dataset)
dataset_label = np.array(dataset_label)

# 학습 데이터세트의 크기를 변환합니다.
dataset = dataset/255.
dataset = dataset.reshape(-1, dataset.shape[1], dataset.shape[2], 1)
```

```python
# 테스트 데이터세트도 동일한 과정을 거칩니다.
test_dataset = []
for test_file in test_dataset_csv['file_name']:
    image_array = np.array(Image.open(test_dataset_dir + test_file))
    test_dataset.append(image_array)
    
test_dataset = np.array(test_dataset)
```

```python
# 테스트 데이터 크기 변환
test_dataset = test_dataset/255.
test_dataset = test_dataset.reshape(-1, test_dataset.shape[1], test_dataset.shape[2], 1)
```

```python
x_train, x_val, y_train, y_val = train_test_split(dataset, dataset_label, test_size = 0.2, stratify = dataset_label)
```

```python
print("Train Dataset Shape: ", x_train.shape)                         # (4000, 28, 28, 1)
print("Train Dataset Label Shape: ", y_val.shape)             # (4000, ) 
print("Validation Dataset Shape: ", x_val.shape)               # (1000, 28, 28, 1)
print("Validation Dataset Label Shape: ", y_val.shape)   # (1000, )
print("Test Dataset Shape: ", test_dataset.shape)                           # (5000, 28, 28, 1)
```

```python
y_train = to_categorical(y_train)
y_val = to_categorical(y_val)
y_test = to_categorical(dataset_label)

display(y_train.shape,y_val.shape, y_test.shape)
```

### NN 모델링

```python
model = Sequential([Input(shape=(28, 28), name = 'input'),
                    Flatten(input_shape=[28, 28], name = 'flatten'),
                    Dense(128, activation = 'relu', name = 'dense1'),
                    Dense(64, activation = 'relu', name = 'dense2'),
                    Dense(32, activation = 'relu', name = 'dense3'),
                    Dense(10, activation = 'softmax', name = 'output')])
```

```python
model.compile(optimizer=Adam(learning_rate=0.01),
              loss="categorical_crossentropy",
              metrics=["accuracy"]
              )
```

```python
print('Logistic Reg model with softmax\n')
# history = model.fit(x_train, y_train, 
#                     epochs=30)

early_stop = EarlyStopping(monitor='val_loss', patience=10)
history = model.fit(x_train, y_train, 
                    epochs=100,
                    batch_size = 64,callbacks=[early_stop],
                    validation_data = (x_val, y_val))
```

### 학습 결과 시각화

```python
history.history.keys()
```

```python
history_dict = history.history

loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs =range(1, len(loss) +1)
fig = plt.figure(figsize=(12,6))

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(epochs, loss, color='blue', label ='train_loss')
ax1.plot(epochs, val_loss, color='red', label='val_loss')
ax1.set_title('Train and Validation loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.grid()
ax1.legend()


accuracy = history_dict['accuracy']
val_accuracy = history_dict['val_accuracy']

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(epochs, accuracy, color='blue', label ='train_acc')
ax2.plot(epochs, val_accuracy, color='red', label='val_acc')
ax2.set_title('Train and Validation acc')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('acc')
ax2.grid()
ax2.legend()

plt.show()
```

### 모델 평가 (1)

```python
model.evaluate(x_val, y_val)
```
